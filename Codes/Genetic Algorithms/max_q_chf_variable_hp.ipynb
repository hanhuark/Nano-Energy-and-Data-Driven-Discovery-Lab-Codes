```python
%matplotlib inline

import random
import math
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import matlab.engine

eng = matlab.engine.start_matlab()
```

```python
def generate_population(size, x_min, y_min, z_min, x_max, y_max, z_max):
    population = []
    for i in range(size):
        individual = generate_individual(x_min, y_min, z_min, x_max, y_max, z_max)
        
        check = apply_function(individual)
        
        if math.isnan(check) == True:
            individual = generate_individual(x_min, y_min, z_min, x_max, y_max, z_max)
        else:
            population.append(individual)
    print('popsize: ', len(population))        
    return population

def generate_individual(x_min, y_min, z_min, x_max, y_max, z_max):
    individual = {
        "x": random.uniform(x_min, x_max),
        "y": random.uniform(y_min, y_max),
        "z": random.uniform(z_min, z_max)
    }
    return individual

def apply_function(individual):
    x = individual["x"]
    y = individual["y"]
    z = individual["z"]
    out = eng.q_chf_cal(x, y, z)
    #print('x: ', x, 'y: ', y, 'out: ', out)
    return out

def choice_by_roulette(sorted_population, fitness_sum):
    draw = random.uniform(0, 1)
    
    accumulated = 0
    for individual in sorted_population:
        fitness = 100 / apply_function(individual)
        probability = fitness / fitness_sum
        accumulated += probability
        
        if draw <= accumulated:
            return individual

def sort_population(population):
    return sorted(population, key=apply_function, reverse=True)

def crossover(individual_a, individual_b):
    xa = individual_a["x"]
    ya = individual_a["y"]
    za = individual_a["z"]
    
    xb = individual_b["x"]
    yb = individual_b["y"]
    zb = individual_b["z"]
    
    return {"x": (xa + xb)/2, "y": (ya + yb)/2, "z": (za + zb)/2}

def mutate(individual, x_min, y_min, z_min, x_max, y_max, z_max):
    next_x = individual["x"] + random.uniform(-2e-6, 2e-6)
    next_y = individual["y"] + random.uniform(-0.1, 0.1)
    next_z = individual["z"] + random.uniform(-1.25, 1.25)
    
    #if (next_x <= 0) or (next_y <= 0):
    #    mutate(individual)
    
    # Keeping above 0.1 and below max values:
    next_x = min(max(next_x, x_min), x_max)
    next_y = min(max(next_y, y_min), y_max)
    next_z = min(max(next_z, z_min), z_max)
    
    return {"x": next_x, "y": next_y, "z": next_z}

def make_next_gen(previous_population, x_min, y_min, z_min, x_max, y_max, z_max):
    next_generation = []
    sorted_population = sort_population(previous_population)
    population_size = len(previous_population)
    fitness_sum = 100 / sum(apply_function(individual) for individual in population)
    
    for i in range(population_size):
        first_choice = choice_by_roulette(sorted_population, fitness_sum)
        second_choice = choice_by_roulette(sorted_population, fitness_sum)
        
        individual = crossover(first_choice, second_choice)
        individual = mutate(individual, x_min, y_min, z_min, x_max, y_max, z_max)
        next_generation.append(individual)
    print('popsize: ', len(next_generation))
    return next_generation
```
