Note that the directories change based on the dataset.

```python
import cv2
import ntpath
import os
```

```python
def save_to_jpg(video_dir, classname):
    vidcap = cv2.VideoCapture(video_dir)
    success,image = vidcap.read()
    count = 0
    while success:
        vidcap.set(1, (count*10))
        filename = "jpg_DS3/" + classname + "/"
        filename = filename + path_leaf(video_dir)
        filename = filename + "frame" + str(count*10) + ".jpg"
        cv2.imwrite(filename, image)     
        success,image = vidcap.read()
        print('Read a new frame: ', success)
        count += 1

def path_leaf(path):
    head, tail = ntpath.split(path)
    return tail or ntpath.basename(head)

def create_data(input_dir):
    classes_list = os.listdir(input_dir)
    for c in classes_list:
        if c != "Class 1, small":
            print(c)
            file_dir = os.path.join(input_dir, c)
            files = os.listdir(file_dir)
            for f in files:
                full_dir = os.path.join(file_dir, f)
                save_to_jpg(full_dir, c)
                pass
        else:
            pass
```
